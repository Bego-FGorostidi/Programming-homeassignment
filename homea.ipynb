### Step 2&3: Localization & Visualization 

```python

import glob 
import nilearn as nil 
from nilearn import plotting, image 

#access location of MRI data 
data = glob.glob("*.nii")  

#create loop for data file 
for file in data: 
    if "anatomical" in file:
        anatomical_data = file 
    if "uniformity" in file:
        functional_data = file 

#This is a useful way of getting your files! Using a for loop to name your files is really effcient and smart. 

#create image from file
functional_image= image.load_img(functional_data)
anatomical_image= image.load_img(anatomical_data) 

#I dont fully understand why you need to load the data as an image, I thought loading the data would work already? interesting though
#Another possible way you could do this is getting the file path and then plotting that directly on plotting.plot_stat_map(your functional data file, bg_img=your anatomical data file)

#create stacked image 
nil.plotting.plot_stat_map(
    stat_map_img=functional_image, 
    bg_img=anatomical_image, 
    title= ("Brain Areas Associated With Psychosis"), 
    cmap="RdPu",
    vmin=(0.1), 
    cut_coords=(17, 4, 8)) 

plotting.show()

#vVery cute figure! :-)
```

 ### Step 4: Histogram 

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt  

#Load the functional data 
img = nib.load(functional_data)
functional_predata = img.get_fdata()

#Extract positive values 
functional_hdata = functional_predata[functional_predata > 0]

#Plot histogram 
plt.figure(figsize=(8, 5))

#Costumize
plt.figure(figsize=(8, 5))
plt.hist(functional_hdata, bins=50, color='lightgreen', edgecolor='black')

#I believe you can plot the histogram directly,you might not need to plt.figure first
plt.title("Distribution Of Positive Values in Psychosis", fontsize=14)
plt.xlabel("Voxel Intensity (Positive Values)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()

#Great work! your figures look amazing. Maybe consider adding some more descriptions of your code?
```
